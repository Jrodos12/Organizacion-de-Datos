#Obtenga la matriz de distancias euclídeas para todos los contenidos que están en Marte. ¿Cuáles son los dos contenidos que están a menor distancia? (⭐⭐)

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
```

```python
df_geo = pd.read_csv("drive/MyDrive/geo_tags.csv")
```

Me quedo con todos los contenidos en marte

```python
df_marte = df_geo.loc[df_geo["gt_globe"] == "mars"].set_index("gt_id")
df_marte
```

```python
def dist(punto, df):
  return np.sqrt(  (df_marte["gt_lat"] - punto["gt_lat"])**2 + (df_marte["gt_lon"] - punto["gt_lon"])**2)
```

```python
matriz_distancias = (df_marte.apply(lambda x: dist(x,df_marte) , axis=1 ))
```

Obtengo la matriz de distancias

```python
matriz_distancias
```

```python
distancia_minima = matriz_distancias[matriz_distancias > 0].min(axis=1).min()
```

obtengo la distancia minima

```python
distancia_minima
```

busco el id, de los 2 contenidos con distancia minima

```python
serie = matriz_distancias[matriz_distancias > 0].min(axis=1) == distancia_minima
```

```python
serie.loc[serie.values == True]
```
