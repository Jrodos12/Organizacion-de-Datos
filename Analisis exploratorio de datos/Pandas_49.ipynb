```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
```

```python
df_lenguajes = pd.read_csv("drive/MyDrive/languages.csv")
df_contents = pd.read_csv("drive/MyDrive/contents.csv")
```

Renombro las columnas para mas comodidad, manejo los nans y transformo a todos los usuarios en si saben el idioma (1) o no (0)

```python
#me quedo solo con las filas que no tengan revisor_id nulo, ya que no es posible relacionar esos usuarios con df_lenguajes
df_lenguajes = df_lenguajes.rename(columns= {"babel_user":"usuario", "babel_lang" : "idioma","babel_level": "nivel_idioma"})
df_lenguajes["sabe_idioma"] = df_lenguajes["nivel_idioma"].map(lambda x: 0 if x == "0" else 1)
df_contents = df_contents.loc[df_contents["revisor_id"].notnull()]
```

#Necesito a los usuarios que sepan el idioma y que hayan hecho una edicion del contendio

```python
df_contents.rename(columns={"revisor_id":"usuario"}, inplace=True)
```

```python
union = pd.merge(df_lenguajes.loc[df_lenguajes["sabe_idioma"] == 1 ], df_contents.loc[ :,["usuario","revisor_comment"] ], on="usuario")
```

```python
def contar(x):
  return(x.isna().sum())
```

```python
df = union.groupby(["idioma"]).agg( {"usuario": "count", "revisor_comment": contar})
```

```python
(100 * df["revisor_comment"]) / df["usuario"]
```
