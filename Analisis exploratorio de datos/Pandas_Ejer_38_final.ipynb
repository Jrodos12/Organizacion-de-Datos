#Obtenga la matriz de correlación para saber idiomas distintos considerando que un usuario sabe un idioma si indicó un nivel de 1 o superior (⭐⭐)

```python
import pandas as pd
import numpy as np
dataframeLenguajes = pd.read_csv("drive/MyDrive/languages.csv")
#cambio los nombre para mas comodidad
dataframeLenguajes = dataframeLenguajes.rename(columns= {"babel_user":"usuario", "babel_lang" : "idioma","babel_level": "nivel_idioma"})
#chequeo la existencia de Nan
dataframeLenguajes.isnull().sum()
```

observo 2 situaciones usuarios con nivel 0 de un idioma, es decir que no sabe el idioma y usuarios con nivel N (nativos) es decir saben el idioma.
Asi mismo observo que 1,2,3,4,5 son equivalentes a saber el idioma

```python
dataframeLenguajes["nivel_idioma"].unique()
```

```python
dataframeLenguajes["sabe_idioma"] = dataframeLenguajes["nivel_idioma"].map(lambda x: 0 if x == "0" else 1)
```

```python
dataframeLenguajes["sabe_idioma"].unique()
```

```python
(dataframeLenguajes["idioma"].isna()).sum()
```

existen idiomas con nans, tomo la decision de tratar dichos nans como lenguajes "otros"

```python
dataframeLenguajes["idioma"].fillna("otros", inplace=True)
```

```python
(dataframeLenguajes["usuario"].isna()).sum()
```

```python
lenguajes_pivot = dataframeLenguajes.pivot("usuario","idioma","sabe_idioma")
```

```python
lenguajes_pivot.fillna(0, inplace=True)
```

```python
lenguajes_pivot
```

```python
#preguntarle a gian, a que se refiere con los numeros
lenguajes_pivot.corr()
```
